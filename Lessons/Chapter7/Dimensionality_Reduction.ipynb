| **Term**         | **Formula**                        | **Dimensions**                | **Meaning**                                      |
|------------------|------------------------------------|-------------------------------|--------------------------------------------------|
| **Data Matrix**  | $X$                                | $n \times d$                  | $n$ samples, $d$ features                        |
| **SVD**          | $X = U \Sigma V^T$                 | $U$: $n \times r$ <br> $\Sigma$: $r \times r$ <br> $V$: $d \times r$ | Factorization of $X$ into orthogonal matrices and singular values |
| **$U$**          | Left singular vectors               | $n \times r$                  | Orthonormal basis for sample space               |
| **$\Sigma$**     | Singular values (diagonal matrix)  | $r \times r$                  | Strength of each principal direction             |
| **$V$**          | Right singular vectors              | $d \times r$                  | Orthonormal basis for feature space              |
| **$r$**          | Rank of $X$                        | $r = \text{rank}(X)$          | Number of nonzero singular values                |
| **Low-rank approx** | $X_k = U_k \Sigma_k V_k^T$      | $k < r$                       | Best rank-$k$ approximation (dimensionality reduction) |
