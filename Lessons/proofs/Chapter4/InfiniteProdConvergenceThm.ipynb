# Infinite Product Convergence Theorem

$$
\text{Let } (a_n)_{n=1}^\infty \text{ be a sequence with } 0 < a_n < 1 \text{ for all } n.
$$

$$
\text{If there exists } \varepsilon > 0 \text{ such that } a_n \le 1 - \varepsilon \text{ for infinitely many } n,
$$

$$
\text{then } \prod_{n=1}^\infty a_n = 0.
$$


$\textbf{Proof}$

$
\text{Let } I = \{ n \in \mathbb{N} \mid a_n \le 1 - \varepsilon \}. \quad \text{By the hypothesis, } I \text{ is an infinite set.} \\ \text{Define the partial products } P_N = \prod_{n=1}^N a_n. $

$$
\text{Since } 0 < a_n < 1 \text{ for all } n, \text{ the sequence } (P_N) \text{ is strictly decreasing and bounded below by 0.}
$$

$$
\text{We can separate the terms of the product into those indices in } I \text{ and those not in } I:
$$

$$
P_N = \left( \prod_{\substack{n=1 \\ n \in I}}^N a_n \right) \left( \prod_{\substack{n=1 \\ n \notin I}}^N a_n \right).
$$

$\text{For } n \notin I, \text{ we have } a_n < 1, \text{ so } \prod_{n \notin I} a_n < 1. \\ \text{For } n \in I, \text{ we have } a_n \le 1 - \varepsilon. \\ \text{Let } k(N) = |I \cap \{1, \dots, N\}| \text{ be the count of indices in } I \text{ up to } N.$

$$
\text{Then, } P_N \le (1 - \varepsilon)^{k(N)} \cdot 1.
$$

$$
\text{Since } I \text{ is infinite, } \lim_{N \to \infty} k(N) = \infty.
$$

$$
\text{Because } 0 < \varepsilon < 1, \text{ we have } 0 < 1 - \varepsilon < 1, \text{ which implies:}
$$

$$
\lim_{N \to \infty} (1 - \varepsilon)^{k(N)} = 0.
$$

$$
\text{By the Squeeze Theorem, since } 0 < P_N \le (1 - \varepsilon)^{k(N)}, \text{ we conclude:}
$$

$$
\lim_{N \to \infty} P_N = 0 \implies \prod_{n=1}^\infty a_n = 0.
$$

$$
\blacksquare
$$
