# Log Function and Information Preservation

## Preface 
This is a proof that's taught early on in most mathematics curricula and can reasonably be assumed as prior knowledge for many readers. <br>
However, since this function arises repeatedly across mathematics, probability and machine learning, it's useful to have the argument written down explicitely. <br>
Additionally, it serves as a clear example of what it means (in a mathematical sense) for a transformation to preserve information, that is for no information to be lost.

#### Intuition for no information to be lost

- If we were provided with a finite set $\{a_1, a_2, \dots, a_n\} \subset (0,\infty)$, of values the log function would be able to map it so that each value in the set is uniquely
  defined - **Uniqueness is preserved**

- If instead we were provided with an ordered set $ ( a_1, a_2, \dots a_n\ )$, the log function would be able to map it so that the relative position between each element in the
  co-domain is maintained - **Order is preserved**

- Finally if we had an infinite set of values $\{a_i\}_{i=1}^{\infty} \subset (0,\infty)$ which has a perticular characteristics as we follow these values to infinity, the log function is able retain these
  characteristics - **Limit behaviour are preserved**


1. $\text{To prove that order is preserved means proving that the function is Strictly monotonically increasing:}$ 
   
$$ 
\forall \ x, y \gt 0 \quad x < y \quad \Longrightarrow \quad \log x < \log y 
$$

$$\textbf{Proof.}$$

$$
a = \log x, \qquad b = \log y \quad  \text{by definition } \quad x = e^a, \qquad y = e^b \\ \text{ We know that } a < b \quad \Longrightarrow \quad e^a \lt e^b. \\ x < y \quad \Longleftrightarrow \quad e^a < e^b \\ \text{ Since } e^x \text{ is stricly increasing then } a < b \text{ thus } \log x < \log y. \\ \blacksquare$$

2. $\text{To prove uniqueness essentially breaks down to proving bijectivity of the function which is injectivity and surjectivity:}$

$$ \text{Injectivity: } \log x = \log y \Longrightarrow \ x = y: \quad \text{ this is direct by (1)} \quad \blacksquare $$

$$ \text{Surjetivity: } \forall t \in \mathbb{R} \quad \exist x \gt 0 | \log(x) = t: \quad x =e^t  \Longrightarrow \log x = \log(e^t) = t \quad \blacksquare$$

1. $\text{To prove that the function doesn't affect the characteristics of the domain set at the limit we prove continuity}$

$$ \forall x_0 \in (0, \infty) \quad \lim_{x \to x_0} \log(x) = \log(x_0) $$


$$\textbf{Proof.}$$

Let $x_0 \in (0,\infty)$ be fixed. We must show that for every $\varepsilon > 0$ there exists $\delta > 0$ such that
$$0 < |x - x_0| < \delta \quad \Longrightarrow \quad |\log x - \log x_0| < \varepsilon.$$

Using the logarithmic identity, for all $x > 0$,
$$|\log x - \log x_0| = \left| \log\!\left(\frac{x}{x_0}\right) \right|.$$

Since $\log$ is continuous at $1$ and $\log(1)=0$, there exists $\eta > 0$ such that
$$|y - 1| < \eta \quad \Longrightarrow \quad |\log y| < \varepsilon.$$

Now observe that
$$\left| \frac{x}{x_0} - 1 \right| = \frac{|x - x_0|}{x_0}.$$

Therefore, if $|x - x_0| < \eta x_0$, then
$$\left| \frac{x}{x_0} - 1 \right| < \eta \quad \Longrightarrow \quad |\log x - \log x_0| < \varepsilon.$$

Choose $\delta = \eta x_0$. Then for all $x$ such that $0 < |x - x_0| < \delta$, we have $|\log x - \log x_0| < \varepsilon$.

Hence,
$$\boxed{\lim_{x \to x_0} \log(x) = \log(x_0)} \quad \blacksquare$$
