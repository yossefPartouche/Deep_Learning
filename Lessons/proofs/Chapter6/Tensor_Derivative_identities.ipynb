$$\boxed{\frac{\partial a^T X b}{\partial X} = ab^T \in \mathbb{R}^{m \times n}}$$


$$a^T X b = \sum_{i=1}^m \sum_{j=1}^n a_i X_{ij} b_j \in \mathbb{R}$$

$$\frac{\partial}{\partial X_{kl}} (a^T X b) = \frac{\partial}{\partial X_{kl}} \left( \sum_{i=1}^m \sum_{j=1}^n a_i X_{ij} b_j \right) = a_kb_l$$

$$ \therefore \frac{\partial a^T X b}{\partial X} = ab^T $$

### Example 

$$
a = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix}, \quad
M = \begin{bmatrix}
    M_{11} & M_{12} \\
    M_{21} & M_{22} \\
    M_{31} & M_{32}
\end{bmatrix}, \quad
b = \begin{bmatrix} b_1 \\ b_2 \end{bmatrix}
$$

Expand $a^T M b$:

$$
a^T M b =
\begin{bmatrix} a_1 & a_2 & a_3 \end{bmatrix}
\begin{bmatrix}
    M_{11} & M_{12} \\
    M_{21} & M_{22} \\
    M_{31} & M_{32}
\end{bmatrix}
\begin{bmatrix} b_1 \\ b_2 \end{bmatrix}
$$

This equals:

$$
\begin{align}
&= a_1(M_{11}b_1 + M_{12}b_2) + a_2(M_{21}b_1 + M_{22}b_2) + a_3(M_{31}b_1 + M_{32}b_2) \\
&= \sum_{i=1}^3 \sum_{j=1}^2 a_i M_{ij} b_j
\end{align}
$$

Take the partial derivative with respect to $M_{ij}$:

$$\frac{\partial}{\partial M_{ij}} (a^T M b) = a_i b_j$$

$$
\begin{align*}
a^T M b &= a_1(M_{11}b_1 + M_{12}b_2) + a_2(M_{21}b_1 + M_{22}b_2) + a_3(M_{31}b_1 + M_{32}b_2) \\
        &= a_1 M_{11} b_1 + a_1 M_{12} b_2 + a_2 M_{21} b_1 + a_2 M_{22} b_2 + a_3 M_{31} b_1 + a_3 M_{32} b_2
\end{align*}
$$

Now, compute the partial derivatives for each entry of $M$:

<div align="center">

| Entry      | Partial Derivative                                 | Result    |
|------------|----------------------------------------------------|-----------|
| $M_{11}$   | $\frac{\partial}{\partial M_{11}} (\textcolor{gold}{a_1 M_{11} b_1} + a_1 M_{12} b_2 + a_2 M_{21} b_1 + a_2 M_{22} b_2 + a_3 M_{31} b_1 + a_3 M_{32} b_2)$       | $a_1 b_1$ |
| $M_{12}$   | $\frac{\partial}{\partial M_{12}} (a_1 M_{11} b_1 + \textcolor{gold}{a_1 M_{12} b_2} + a_2 M_{21} b_1 + a_2 M_{22} b_2 + a_3 M_{31} b_1 + a_3 M_{32} b_2)$       | $a_1 b_2$ |
| $M_{21}$   | $\frac{\partial}{\partial M_{21}} (a_1 M_{11} b_1 + a_1 M_{12} b_2 + \textcolor{gold}{a_2 M_{21} b_1} + a_2 M_{22} b_2 + a_3 M_{31} b_1 + a_3 M_{32} b_2)$       | $a_2 b_1$ |
| $M_{22}$   | $\frac{\partial}{\partial M_{22}} (a_1 M_{11} b_1 + a_1 M_{12} b_2 + a_2 M_{21} b_1 + \textcolor{gold}{a_2 M_{22} b_2} + a_3 M_{31} b_1 + a_3 M_{32} b_2)$       | $a_2 b_2$ |
| $M_{31}$   | $\frac{\partial}{\partial M_{31}} (a_1 M_{11} b_1 + a_1 M_{12} b_2 + a_2 M_{21} b_1 + a_2 M_{22} b_2 + \textcolor{gold}{a_3 M_{31} b_1} + a_3 M_{32} b_2)$       | $a_3 b_1$ |
| $M_{32}$   | $\frac{\partial}{\partial M_{32}} (a_1 M_{11} b_1 + a_1 M_{12} b_2 + a_2 M_{21} b_1 + a_2 M_{22} b_2 + a_3 M_{31} b_1 + \textcolor{gold}{a_3 M_{32} b_2})$       | $a_3 b_2$ |

</div>

So, the gradient matrix is:

$$
\frac{\partial a^T M b}{\partial M} =
\begin{bmatrix}
    a_1 b_1 & a_1 b_2 \\
    a_2 b_1 & a_2 b_2 \\
    a_3 b_1 & a_3 b_2
\end{bmatrix}
= a b^T
$$
